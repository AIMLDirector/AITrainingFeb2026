```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
sentence1 = 'Sentiment analysis uses natural language processing (NLP) and machine learning (ML) technologies to train computer software to analyze and interpret text in a way similar to humans. The software uses one of two approaches, rule-based or ML—or a combination of the two known as hybrid.Each approach has its strengths and weaknesses'
```

```python
tokenized_word = word_tokenize(sentence1.lower())
print(tokenized_word)
```

```python
sent_tokenize(sentence1)
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
print(stop_words)
```

```python
filtered_tokens = [word for word in tokenized_word if word not in stop_words]
print(filtered_tokens)
```

```python
import string
cleaned_tokens = [word for word in filtered_tokens if word not in string.punctuation]
print(cleaned_tokens)
            
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(sentence1.lower())
print(tokens)
```

Stemming :
1. Fast/ rule based/non linguistical accurate

running - run
studies - studi
playing - play 

```python
from nltk.stem import PorterStemmer, SnowballStemmer, LancasterStemmer
porter = PorterStemmer()
snowball = SnowballStemmer(language='english')
lancaster = LancasterStemmer()
```

```python
for i in tokens:
    print(porter.stem(i))
    
    
```

```python
for i in tokens:
    print(snowball.stem(i))
```

```python
for i in tokens:
    print(lancaster.stem(i))
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
for i in tokens:
    print(lemmatizer.lemmatize(i))     # NOUN 
```

```python
print(lemmatizer.lemmatize("processing"))
```

```python
print(lemmatizer.lemmatize("processing", pos="v"))
```

```python
print(lemmatizer.lemmatize("processing", pos="a"))
```

```python
print(lemmatizer.lemmatize("running", pos="v"))
```

```python
print(lemmatizer.lemmatize("running"))
```

```python
from nltk import pos_tag, ne_chunk
taged_token = pos_tag(tokens)
print(taged_token)
```

```python
NER_token = ne_chunk(taged_token)
print(NER_token)
```

```python
tokens = word_tokenize(sentence)
pos_token = pos_tag(tokens)
Ner_token = ne_chunk(pos_token)
print(Ner_token)
```

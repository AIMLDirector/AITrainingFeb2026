```python
import textblob
```

```python
sentence1 = "Find in the list below the best Sentinelone resellers or channel partners that are currently on our platform to help you with implementation"
```

```python
from textblob import TextBlob
blob = TextBlob(sentence1)
```

```python
print("words:", blob.words)
print("sentence:", blob.sentences)
```

```python
print("POS tagging:", blob.tags)
```

```python
print("sentiment:", blob.sentiment)
```
